### Univariate analysis 
#### This graph shows the diffrence in between the fully paid loan people and charged off people  

```python
tr0 = [
             data[data["loan_status"]== 'Fully Paid']["loan_status"].value_counts().index.values,
             data[data["loan_status"]== 'Fully Paid']["loan_status"].value_counts().values,
            
]

tr1 = [
             data[data["loan_status"]== 'Charged Off']["loan_status"].value_counts().index.values,
            data[data["loan_status"]== 'Charged Off']["loan_status"].value_counts().values,
            
]
plt.figure(figsize =(10, 10))
plt.bar( tr0[0],tr0[1], 0.4, label = 'Fully Paid')
plt.bar( tr1[0],tr1[1], 0.4, label = 'Charged Off')
plt.legend(["Fully_paid", "Charged_Off"])
plt.xlabel("Loan_status")
plt.ylabel("Count")
plt.title( "This is a graph represent the total count of people that have \n paid the loan fully vs those who got charged off",fontsize = 15)
# plt.text(-1, -1, 'This is a graph represent the total count of people that have \n paid the loan fully vs those who got charged off', fontsize = 20)
plt.show()
```

### Bivariate analysis 

```python
df_Fully_paid=data[data["loan_status"]== 'Fully Paid']["home_ownership"].values.tolist()
df_Charged_Off=data[data["loan_status"]== 'Charged Off']["home_ownership"].values.tolist()
```

```python
plt.figure(figsize =(10, 10))
plt.hist(df_Fully_paid,linewidth=10)
plt.hist(df_Charged_Off,linewidth=1)
plt.legend(["Fully_paid", "Charged_Off"])
plt.xlabel("Home_ownership")
plt.ylabel("Loan_status_Count")
plt.title( "This is a graph represent how the diffrent type of renting criteria can be a deriving factor loan payment and charged off ",fontsize = 15)
plt.show()
```

```python
df_Fully_paid_emp_length=data[data["loan_status"]== 'Fully Paid']["emp_length"].values.tolist()
df_Charged_Off_emp_length=data[data["loan_status"]== 'Charged Off']["emp_length"].values.tolist()
plt.figure(figsize =(20, 20))
plt.hist(df_Fully_paid_emp_length,linewidth=10)
plt.hist(df_Charged_Off_emp_length,linewidth=1)
plt.legend(["Fully_paid", "Charged_Off"])
plt.xlabel("Emp_length")
plt.ylabel("Loan_status_Count")
plt.title( "This is a graph represent how the employee length is a factor to the contribution of fully payment and charged off ",fontsize = 15)
plt.show()
```

```python
df_Fully_paid_annual_inc=data[data["loan_status"]== 'Fully Paid']["annual_inc"].values.tolist()
df_Charged_Off_annual_inc=data[data["loan_status"]== 'Charged Off']["annual_inc"].values.tolist()
plt.figure(figsize =(20, 20))
plt.hist(df_Fully_paid_annual_inc,linewidth=10)
plt.hist(df_Charged_Off_annual_inc,linewidth=1)
plt.legend(["Fully_paid", "Charged_Off"])
plt.xlabel("Annual_inc")
plt.ylabel("Loan_status_number_Count")
plt.title( "This is a graph represent how the annual income can be used a derving factor of laon payment ",fontsize = 15)
plt.show()
```
